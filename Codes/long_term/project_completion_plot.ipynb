```python
#create function to categorize projects in the categories
def completion(x):
    if x == 0:
        return 'Not Observed'
    elif x < 100:
        return 'Partially Observed'
    elif x == 100:
        return 'Completed'
    else:
        return pd.NaT
```

```python
df['Completion'] = df.compl.apply(completion) #apply function
```

```python
count_by_grade_comp = df.groupby(['grade', 'Completion']).count() #get amount of projects in categories and grade
```

```python
count_by_grade = df.groupby(['grade']).count() #get amount of projects per grade
```

```python
count_by_grade_comp = count_by_grade_comp.reset_index().set_index('grade') #set grade as index
```

```python
count_by_grade_comp['Percentage'] = 0 #create Percentage column
```

```python
#calculate percetages per grade
count_by_grade_comp.loc['A', 'Percentage'] = 100 * count_by_grade_comp.loc['A', 'compl'] / count_by_grade.loc['A', 'compl']
count_by_grade_comp.loc['B', 'Percentage'] = 100 * count_by_grade_comp.loc['B', 'compl'] / count_by_grade.loc['B', 'compl']
count_by_grade_comp.loc['C', 'Percentage'] = 100 * count_by_grade_comp.loc['C', 'compl'] / count_by_grade.loc['C', 'compl']
```

```python
#create palette for plot
custom_palette = {'Completed': '#5799C7', 'Partially Observed': '#C3D5EE', 'Not Observed': '#FF9F4B'}
```

```python
#create plot
px.bar(count_by_grade_comp.reset_index(), 
             x='grade', 
             y='Percentage', 
             color='Completion',
             category_orders={'grade': ['A', 'B', 'C'], 'completion': ['Completed', 'Partially Observed', 'Not Observed']},
             color_discrete_map=custom_palette,
             #title=titulo,
             text_auto='.2s')
```
